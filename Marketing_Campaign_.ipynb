2. Data Cleaning and Transformation

```python
#IQR Calculation to find out if there exists any outliers
for col in columns: #loop through each col
    Q1 = df[col].quantile(0.25)#calculate 25th percentile 
    Q3 = df[col].quantile(0.75)#calculate 75th percentile 
    IQR = Q3 - Q1 #Compute the outliers
    lower_bound = Q1 - 1.5*IQR #determine lower-bound or upper bound values outside these are considered outliers
    upper_bound = Q3 + 1.5*IQR
    print("Outliers in column",col)
    outliers = df[(df[col]< lower_bound)|(df[col]>upper_bound)]#filter out outliers
    print(outliers)
```

Using IQR calculation we can see that every column is giving output as empty dataframe which indicates there exists no any outliers,so there is no need to treat outliers.

```python
#Data Transformation: 
df_copy = df.copy()#create copy of df which doesnot affect original dataset 
df_copy.head()#display the first five rows
```

```python
#Indentify unique values in categorical columns
categorical_columns = df_copy.select_dtypes(include=['object']).columns #select column with datatype object
for column in categorical_columns:#loop through categorical columns
    values = df_copy[column].unique()#select the unique values
    print(f'{column}:{values}') #Display the result
    print()
```

From above information we can see the unique values in categorical columns where Advertising platform and Advertising tool are confidential so we will be removing those columns.

```python
#Dropping the confidential column including customer id
df_copy.drop(['CustomerID','AdvertisingPlatform','AdvertisingTool'],axis=1,inplace=True)
```

```python
#Identify unique values in categorical columns after dropping confidential columns
categorical_columns = df_copy.select_dtypes(include=['object']).columns
for column in categorical_columns:
    values = df_copy[column].unique()
    print(f'{column}:{values}')
    print()
```

```python
#Encode categorical variables using LabelEncoder
from sklearn.preprocessing import LabelEncoder #Importing labelEncoder from scikit learn
label_encoders = {} #dictionary to store labelEncoder objects for each categorical column
for column in categorical_columns: #Loop through each categorical column
    le = LabelEncoder()  # Instantiate LabelEncoder for current column
    df_copy[column] = le.fit_transform(df[column])#fit and transform categorical column into numerical column
    label_encoders[column] = le #Store the LabelEncoder Object

for column in categorical_columns:#Loop through each categorical column to print unique transformed values
    print(column, df_copy[column].unique()) #Print the unique encoded values 
```

```python
#Indentify unique values in numerical columns
numerical_columns = df_copy.select_dtypes(include=['int64','float64']).columns #select column with datatype 'int64' and 'float64' 
for column in numerical_columns: #loop through numerical columns
    values = df_copy[column].unique()#select unique value from each column
    print(f'{column}:{values}')#print the result
    print()
```

```python
from sklearn.preprocessing import MinMaxScaler #importing MinMaxScaler from sklearn.preprocessing
min_max = MinMaxScaler() #Create an instance a MinMaxScaler object to scale numerical data
df_copy[numerical_columns]=min_max.fit_transform(df_copy[numerical_columns])#Apply MinMax scaling to numerical columns in dataframe
print(df_copy[numerical_columns])#print the result
```

```python
from sklearn.preprocessing import StandardScaler #Importing StandardScaler from scikit learn
scaler = StandardScaler() #Create an instance a StandardScaler object to scale numerical data
df_copy[numerical_columns] = scaler.fit_transform(df_copy[numerical_columns]) #Apply Standard scaling to numerical columns in dataframe
print(df_copy[numerical_columns])#print the result
```

Create new features(cost per acquisition, click-through rate, 
campaign duration)

```python
#creating new column CostPerAcquisition
df['CostPerAcquisition'] = df['AdSpend'] /df['Conversion']# divides the amount spend on ads(AdSpend) by the rate of conversionrate and the result is stored in new column CostPerAcquisition

#creating new column Click_throughrate
#Calculate clickthroughrate percentage by dividing the sum of ClickThroughRate values by total number of rows in the dataframe and multiplying by 100
df['Click_throughrate'] = df['ClickThroughRate'].mean()*100


#create new column campaign duration
#Calculate campaign_duration by dividing Timeonsite by PagesPerVisit 
df['Campaign_Duration'] = df['TimeOnSite'] / df['PagesPerVisit']

df.head()
```
